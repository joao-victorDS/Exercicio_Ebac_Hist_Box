```python
# Importar Bibliotecas
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Metodo do notebook 
%matplotlib inline
# Style 
sns.set_style("whitegrid")
```

```python
# Importar dados em formato dataframe
df_sus = pd.read_csv("./datasus_nascidos_vivos.csv")

df_sus.head()
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
# Modificar tamanho da figura
plt.rc("figure", figsize=(7,6))

# Plotar grafico do tipo box para variavel "IDADEMAE"
ax = sns.boxplot(data=df_sus, y="IDADEMAE", color="red")

# Alterar padroes esteticos
ax.tick_params(labelsize="large", grid_linestyle="--")
plt.ylabel("Idade (anos)", size=12)
plt.title("Gráfico de Idade da Mãe", size=16)
plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
# Modificar tamanho da figura
plt.rc("figure", figsize=(7,6))

# Plotar grafico estilo box para varivel "PESO"
sns.boxplot(data=df_sus, y="PESO", flierprops={"marker": "_"}, color="orange")

# Alterar padroes esteticos
plt.title("Gráfico de Peso do Bebê", size=16)
plt.ylabel("Peso (g)", size=12)
plt.tick_params(labelsize=12, grid_linestyle="--")
```

####  3) Faça um histograma da idade da mãe

```python
# Plotar grafico estilo hist para varivel "IDADEMAE"
sns.displot(data=df_sus, x="IDADEMAE", bins=20, color="r")

# ALterar padroes esteticos
plt.tick_params(grid_linestyle="--", labelsize=12)
plt.title("Gráfico de Idade da Mãe", size=16)
plt.ylabel("Contagem", size=12)
plt.xlabel("Idade (anos)", size=12)
plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
# Plotar grafico estilo hist para varivel "PESO"
sns.histplot(data=df_sus, x="PESO", bins=30, color="orange")

# Alterar padroes esteticos
plt.title("Gráfico de Peso do Bebê", size=16)
plt.ylabel("Contagem", size=12)
plt.xlabel("Peso (g)", size=12)
plt.tick_params(labelsize=12, grid_linestyle="--")
```
